```python
selected_columns = v[['Age', 'Gender', 'Platform', 'Watch Time', 'Addiction Level']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
train_data, test_data = train_test_split(selected_columns, test_size=0.25, random_state=1)
```

```python
print("Training Data Shape:", train_data.shape)
print("Testing Data Shape:", test_data.shape)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(v['Age'], bins=20, color='blue', alpha=0.7)
plt.title('Age Distribution of Users')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=v)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Platform', data=v)
plt.title('Platform Usage Distribution')
plt.xlabel('Platform')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
train_data_encoded = pd.get_dummies(train_data, drop_first=True)
test_data_encoded = pd.get_dummies(test_data, drop_first=True)
```

```python
X_train = train_data_encoded.drop('Addiction Level', axis=1)
y_train = train_data_encoded['Addiction Level']

X_test = test_data_encoded.drop('Addiction Level', axis=1)
y_test = test_data_encoded['Addiction Level']



model = LogisticRegression(random_state=42, max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the model: {accuracy:.2f}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
report = classification_report(y_test, y_pred)
print("Classification Report:")
print(report)
```
