```python
selected_columns = v[['Age', 'Gender', 'Platform', 'Watch Time', 'Addiction Level']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
train_data, test_data = train_test_split(selected_columns, test_size=0.25, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
train_data_encoded = pd.get_dummies(train_data, drop_first=True)
test_data_encoded = pd.get_dummies(test_data, drop_first=True)
```

```python
X_train = train_data_encoded.drop('Addiction Level', axis=1)
y_train = train_data_encoded['Addiction Level']

X_test = test_data_encoded.drop('Addiction Level', axis=1)
y_test = test_data_encoded['Addiction Level']



model = LogisticRegression(random_state=42, max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)

# Assuming you have 8 classes, replace with your actual class names
class_names = ['Age', 'Gender', 'Location', 'Income', 'Debt', 'Platform', 'Owns Property', 'Demographics']

# Create the DataFrame for the confusion matrix
df_conf_matrix = pd.DataFrame(conf_matrix,
                              index=class_names,
                              columns=class_names)

print("Confusion Matrix:")
print(df_conf_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df_conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Class')
plt.ylabel('Actual Class')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
